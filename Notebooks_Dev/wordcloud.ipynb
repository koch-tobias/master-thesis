```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#convert list to string and generate
my_list = ["LightGBM", "XGBoost", "CatBoost", "KI", "NeuronaleNetze", "Entscheidungsbaum", "GradientBoosting" ]
unique_string=(" ").join(my_list)
wordcloud = WordCloud(width = 1000, height = 500, background_color="rgba(255, 255, 255, 0)", mode="RGBA").generate(unique_string)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, cmap='Blues')
plt.axis("off")
wordcloud.to_file("wordcloud.png")
plt.show()
plt.close()
```
